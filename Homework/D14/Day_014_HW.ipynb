### 請根據不同的 HOUSETYPE_MODE 對 AMT_CREDIT 繪製 Histogram

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = '../../data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
app_train['HOUSETYPE_MODE'] = app_train['HOUSETYPE_MODE'].replace(np.nan, 'none', regex=True)   #把空值取代為 'none'
unique_house_type = list(set(app_train['HOUSETYPE_MODE']))

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    
    #print(app_train.loc[(app_train['HOUSETYPE_MODE'] == unique_house_type[i]), ['AMT_CREDIT', 'HOUSETYPE_MODE']].head(5))
    app_train.loc[(app_train['HOUSETYPE_MODE'] == unique_house_type[i]), 'AMT_CREDIT'].hist()
    
    plt.title(str(unique_house_type[i]))
plt.show() 
```
