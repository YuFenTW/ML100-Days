#### 資料的 row 數以及 column 數

```python
#顯示前 5 筆資料
app_train.head()
```

```python
#找出欄位 NAME_CONTRACT_TYPE 中的唯一值
app_train['NAME_CONTRACT_TYPE'].unique()
```

```python
#只顯示某指定行 (column) 的`值`符合條件的資料列 (row) ==> 數字 (等於)
app_train[app_train['CNT_CHILDREN']==10]
```

```python
#只顯示某指定行 (column) 的`值`符合條件的資料列 (row) ==> 數字 (大於)
app_train[app_train['CNT_CHILDREN']>11]
```

```python
#只顯示某指定行 (column) 的`值`符合條件的資料列 (row) ==> 字串
app_train[app_train['NAME_CONTRACT_TYPE']=='Revolving loans']
```

```python
#資料的 row 數以及 column 數
#output ==> row, column
app_train.shape
```

#### 截取部分資料

```python
#截取前 5 rows, 前 8 columns 
app_train.iloc[:5, :8]
```

```python
#截取 rows (3~5) 共 2 rows, columns (2,6) 共 4 columns ==> 特定範圍
app_train.iloc[3:5, 2:6]
```
