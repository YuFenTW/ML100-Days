# 作業 : (Kaggle)房價預測

```python
# 關閉 FutureWarning 的警告
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# 做完特徵工程前的所有準備 (與前範例相同)
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

%matplotlib inline

data_path = '../../data/2/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
df = df.astype(np.float64)
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
df_trim = df.copy()  # 為了不影響 df, 才能反覆修改 lower & upper 的值
df_trim['1stFlrSF'] = df_trim['1stFlrSF'].clip(330, 2480)
sns.regplot(x = df_trim['1stFlrSF'], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
df_trim = df_trim.astype(np.float64)
train_X = MMEncoder.fit_transform(df_trim)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

## 回答 : 改為 '捨棄離群值' 的方法, 結果會變好, 因為可以避免因為 '調整離群值' 而造成數據會在二極 (上下限) 堆疊的狀況發生

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
df_replace = df.copy()  # 為了不影響 df, 才能反覆修改 lower & upper 的值
keep_indexs = (df_replace['1stFlrSF']> 330) & (df_replace['1stFlrSF']< 2480)

df_replace = df_replace[keep_indexs]

train_Y = train_Y[keep_indexs]
sns.regplot(x = df_replace['GrLivArea'], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
df_replace = df_replace.astype(np.float64)
train_X = MMEncoder.fit_transform(df_replace)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```
