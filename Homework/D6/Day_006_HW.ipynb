# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = '../../data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
dtype_select = app_train.select_dtypes(include=['int64', 'float64'])

numeric_columns = list(app_train.columns[list(app_train.columns.isin(dtype_select))])


# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
print("Numbers of remain columns %d" %len(numeric_columns))

# 檢視這些欄位的數值範圍
for col in numeric_columns:
    app_train.boxplot(col, 
                      flierprops = {'marker':'o','markerfacecolor':'red','color':'black'}, # 设置异常值属性，点的形状、填充色和边框色
                     )
    plt.show()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE
```

```python
# ECDF = 經驗累積函數
def ecdf(data):
    # 1) value_counts : 計算出各個值出現的次數==>series[值(index), 出現次數] 
    # 2) sort_index : 對值進行升序排序==>series[值(index), 出現次數]
    # 3) cumsum : 累計==>[值(index), 每值出現的次數 + 所有比他小的值出現的數次合計 = 累計次數]
    x = data.value_counts().sort_index().cumsum() 

    # 整個 data 的數筆共有多少筆 = 總筆數
    y = len(data) 
    
    # series [值(index), 累計次數] / 總筆數 = series [值(index), 累計次數 / 總筆數]
    # column[0]的值是做 index 用, 所以不能用來作除法運算 
    return  x / y
```

```python
# 最大值離平均與中位數很遠
#print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
cdf = ecdf(app_train['AMT_INCOME_TOTAL'])
plt.plot(list(cdf.index), cdf/cdf.max()) # 線圖 ==> plot(x, y)
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

# 改變 x 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()
```
